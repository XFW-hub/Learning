## 排序的依据

- 排序模型预估点击率、点赞率、收藏率、转发率等多种分数
- 融合预估分数（如加权和）
- 根据融合的分数做排序、截断

## 多目标模型

- 将用户特征、物品特征、统计特征、场景特征拼接输入神经网络
- 神经网络输出向量
- 将向量输入四个神经网络
- 四个神经网络分别输出点击率，点赞率，收藏率，转发率

### 模型的训练

总的损失函数：

$
\sum_{i=1}^4{\alpha _i·CrossEntropy}\left( y_i,p_i \right) 
$

困难：类别不平衡

负样本降采样（down-sampling）

### 预估值校准

真实点击率：$
p_{true\,\,=\,\,\frac{n_+}{n_++n_-}}
$

预估点击率：
$
p_{true\,\,=\,\,\frac{n_+}{n_++\alpha ·n_-}}
$

校准公式：
$
p_{true\,\,}=\,\,\frac{\alpha ·p_{pred}}{\left( 1-p_{pred} \right) +\alpha ·p_{pred}}
$
