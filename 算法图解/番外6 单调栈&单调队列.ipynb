```python
def nearestGeeater(nums:List[int])->Tuple[List[int],List[int]]:
    n = len(nums)
    left = [-1]*n
    st = []
    for i,x in enumerate(nums):
        while st and nums[st[-1]]<=x:
            st.pop()
        if st:
            left[i] = st[-1]
        st.append(i)

    right = [n]*n
    st = []
    for i in range(n-1,-1,-1):
        x = nums[i]
        while st and nums[st[-1]]<=x:
            st.pop()
        if st:
            right[i] = st[-1]
        st.append(i)
    return left,right
```

```python
def maxSlidingWindow(nums:List[int],k:int)->List[int]:
    ans = [0] * (len(nums)-k+1)
    q = deque()

    for i,x in enumerate(nums):
        while q and nums[q[-1]]<=x:
            q.pop()
        q.append(i)

        left = i-k+1
        if q[0]<left:
            q.popleft() ###关键在于这一步 所以要用双端队列
        if left>=0:
            ans[left]=nums[q[0]]
    return ans
```
