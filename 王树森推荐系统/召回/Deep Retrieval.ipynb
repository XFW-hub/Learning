### 索引

1.索引1 物品到路径

一个物品对应多条路径

2.索引2 路径到物品

一条路径对应多个物品



### 预估模型

预估用户对路径的兴趣

path = [a,b,c]

p(a,b,c) = p1(a|x)\*p2(b|a;x)\*p3(c|a,b;x)

最优路径：[a*,b*,c*] = argmaxp(a,b,c|x)

### 线上召回

用户——路径——物品

第一步：给定用户特征，用beam search召回一批路径

第二步：利用路径到物品的索引召回一批物品

第三步：对物品做打分和排序，选出子集
