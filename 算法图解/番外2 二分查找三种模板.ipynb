## 模板1 标准二分查找

```python
def binary_search(nums, target):
    n = len(nums)
    if n==0:return -1
    l,r = 0,n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid]==target:
            return mid
        elif nums[mid]<target:
            l = mid+1
        else:
            r = mid-1
    return -1
```

## 模板2 寻找左边界

```python
def binary_search(nums, target):
    n = len(nums)
    if n==0:return -1
    l,r = 0,n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid]<target:
            l = mid+1
        else:
            r = mid-1
    if l==n:return -1
    return nums[l]#数组中第一个大于等于target的元素
```

## 模板3 寻找右边界

```python
def binary_search(nums, target):
    n = len(nums)
    if n==0:return -1
    l,r = 0,n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid]<=target:
            l = mid+1
        else:
            r = mid-1
    if l<0:return -1
    return nums[r]#数组中第一个小于等于target的元素
```
