## 曝光过滤

看过某个物品，不再把物品曝光给用户

## Bloom Filter

一种数据结构

如果判断为no：一定不在集合中

判断为yes:该物品很可能在集合中



- 把物品集合表征为m维二进制向量
- 每个用户有一个曝光物品的集合，表征为一个向量，需要m bits的存储
- 有k个哈希函数,每个哈希函数把物品ID映射成介于0和m-1之间的整数

### 误伤概率

曝光物品的集合大小为n，二进制向量维度为m，使用k个哈希函数

误伤概率：

$
\delta \,\,\approx \,\,\left( 1-\exp \left( -\frac{kn}{m} \right) \right) ^k
$

最优参数：

$
k\,\,=\,\,1.44·\ln \left( \frac{1}{\delta} \right) 
$

$
m\,\,=\,\,2n·\ln \left( \frac{1}{\delta} \right) 
$
