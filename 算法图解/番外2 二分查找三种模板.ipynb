## 普通模板1 标准二分查找

```python
def binary_search(nums, target):
    n = len(nums)
    if n==0:return -1
    l,r = 0,n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid]==target:
            return mid
        elif nums[mid]<target:
            l = mid+1
        else:
            r = mid-1
    return -1
```

## 普通模板2 寻找左边界

```python
def binary_search(nums, target):
    n = len(nums)
    if n==0:return -1
    l,r = 0,n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid]<target:
            l = mid+1
        else:
            r = mid-1
    if l==n:return -1
    return nums[l]#数组中第一个大于等于target的元素
```

## 普通模板3 寻找右边界

```python
def binary_search(nums, target):
    n = len(nums)
    if n==0:return -1
    l,r = 0,n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid]<=target:
            l = mid+1
        else:
            r = mid-1
    if l<0:return -1
    return nums[r]#数组中第一个小于等于target的元素
```

## 高级模板1 查找条件需要用到右邻居

```python
def binarySearch(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: int
    """
    if len(nums) == 0:
        return -1

    left, right = 0, len(nums)
    while left < right:#保证每一步的查找空间都有至少两个元素
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid
    if left != len(nums) and nums[left] == target:
        return left
    return -1
```
