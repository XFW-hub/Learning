## 组合型回溯 剪枝

### 剪枝

len(path)=m 
需要选k个数
仍需要选d=k-m个数

设当前需要从[1,i]取数，如果i < d,则不需要递归

```python
#枚举选哪个
class solution:
    def combine(self,n:int,k:int) -> List[List[int]]:
        ans = []
        path = []
        def dfs(i):
            d = k-len(path)
            if len(path)==k:
                ans.append(path.copy())
                return
            for j in range(i,d-1,-1):
                path.append(j)
                dfs(j-1)
                path.pop()
        dfs(n)
        return ans
```

## 排列型回溯

```python
#写法1：利用集合
class solution:
    def permuate(self,nums:List[int])->List[List[int]]:
        n = len(nums)
        path = [0] *n
        ans = []
        def dfs(i,s):
            if i==n:
                ans.append(path.copy())
                return
            for x in s:
                path[i] = x
                dfs(i+1,s-{x})
        dfs(0,set(nums))
        return ans

```

```python
#写法2：利用布尔数组
class solution:
    def permuate(self,nums:List[int])->List[List[int]]:
        n = len(nums)
        path = [0] *n
        ans = []
        on_path = [False]*n
        def dfs(i):
            if i==n:
                ans.append(path.copy())
                return
            for j in range(n):
                if not on_path[j]:
                    path[i] = nums[j]
                    on_path[j] = True
                    dfs(i+1)
                    on_path[j] = False
        dfs(0)
        return ans
```
