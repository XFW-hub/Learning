### 推荐系统的链路

1.召回

召回通道：

协同过滤、双塔模型、关注的作者

2.粗排

小规模神经网络

输入：用户特征、物品特征、统计特征

输出：排序分数（点击率预估值、点赞率预估值的加权和）

3.精排

大规模深度学习网络

4.重排

多样性抽样（MMR、DPP）

用规则打散相似笔记

插入广告、运营内容




### A/B测试

1.随机分桶

2.分层实验

解决流量不够用的问题

（1）同层互斥

同层的不同不同实验占不同的桶

ui交uj是空集

（2）不同层正交

每一层独立随机对用户做分桶、每一层独立用100%用户做实验

ui交vj=分数相乘

一个层的一个桶再均匀分为另一个层的某个实验的n个桶

3.Holdout机制

取10%用户做holdout桶，推荐系统使用剩余用户做实验，两者互斥

计算二者归一化后的diff为整个部门的业务指标收益

每个考核周期结束清除Holdout桶，推全实验扩大到100%用户

4.实验推全&反转实验

有的指标立刻收到新策略影响

有的指标有滞后性，需要长期观测

在推全层留一个反转桶、在反转桶保留旧策略
